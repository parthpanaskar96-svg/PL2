**Frequency of characters in a string**

```
def count_char_frequencies(input_text):
  frequency={}
  for char in input_text:
    frequency[char]=frequency.get(char,0)+1
  return frequency
text=input("Enter a string:")
char_counts=count_char_frequencies(text)
print(f"Chatacter frequencies for '{text}'")
for char,count in char_counts.items():
  print(f"{char}:{count}")
```
