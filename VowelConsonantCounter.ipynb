**Vowel and consonant counter**

```
string=input("Enter a string:")
vowel=[]
consonant=[]
for char in string:
  if char.isalpha():
    if char.lower() in 'aeiou':
      vowel.append(char)
    else:
      consonant.append(char)
print(f"---{len(vowel)} Vowels in {string}---")
if vowel:
  print(f"   *{vowel}")
print(f"---{len(consonant)} Consonant in {string}")
if consonant:
  print(f"    *{consonant}")
```
