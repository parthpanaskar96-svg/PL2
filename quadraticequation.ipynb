**Quadratic Equation Solver**

```
import math
a=int(input("Enter the coefficient for x^{2}:"))
b=int(input("Enter the coefficient for x:"))
c=int(input("Enter the coefficient for constant part:"))
discriminant=b**2-4*a*c
if discriminant>0:
  #there are 2 roots which are real and distinct
  root1=(-b+math.sqrt(discriminant))/(2*a)
  root2=(-b-math.sqrt(discriminant))/(2*a)
  print("root1:",root1)
  print("root2:",root2)
elif discriminant==0:
   #there is only one root
   root=-b/(2*a)
   print("root:",root)
else:
    #roots are complex number
    real_part=-b/(2*a)
    imaginary_part=math.sqrt(abs(discriminant))/(2*a)

    print(f"root1:{real_part}+{imaginary_part}i")
    print(f"root2:{real_part}-{imaginary_part}i")



```
