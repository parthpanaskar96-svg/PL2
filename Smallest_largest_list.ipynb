**Finding smallest and largest elemengt from list**

```
list=[]
size=int(input("Enter the size of list: "))

for i in range(size):
  n=int(input("Enter the element: "))
  list.append(n)
print("Smallest element from list is ",min(list))
print("Largest element from list is ",max(list))
```
