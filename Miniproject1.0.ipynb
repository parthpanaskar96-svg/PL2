#**MINI PROJECT 1.0**


**Daily Checklist**

```
checklist=[]
completed_task=[]
incompleted_task=[]
print("Let's create your daily checklist.Enter tasks one by one, type 'done' if finished." )
while True:

    task=input("Enter the task:")
    if task.lower()=='done':
      break
    checklist.append(task)
print("----Review your tasks----")
if not checklist:
  print("No tasks were added for today.")
else:
  for task in checklist:
    while True:
      status=input(f"Did you complete '{task}'?(Yes/No)")
      if status.lower()=='yes':
        completed_task.append(task)
        break
      elif status.lower()=='no':
          incompleted_task.append(task)
          break
      else:
        print("Please enter only Yes or No.")
print("\n----Daily Overview----")
print("\n----Original Checklist----")
if checklist:
  for task in checklist:
    print(f"    - {task}")
else:
  print("  No tasks were originally added.")
print(f"----Completed tasks: {len(completed_task)}----")
if completed_task:
  for task in completed_task:
    print(f"    - {task}")

print(f"----Incompleted tasks: {len(incompleted_task)}----")
if incompleted_task:
  for task in incompleted_task:
    print(f"    - {task}")



print(f"Congratulations!!! You have completed {len(completed_task)} out of {len(checklist)} tasks today. ")
```
